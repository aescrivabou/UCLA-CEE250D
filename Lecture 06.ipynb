# Define simulation policies
## Exercise 3.17 in Loucks and van Beek (2017)

```python
import numpy as np

gage_flow = np.array(range(130))

print(gage_flow)
```

```python
#Initialize variables
env_flow = np.zeros(130)
user1 = np.zeros(130)
user2 = np.zeros(130)
user3 = np.zeros(130)

#Simulate for each possible flow
for f in range(130):
    if f <10:
        env_flow[f] = f
    elif f<30:
        env_flow[f] = 10
        user1[f]= f-10
    elif f<90:
        env_flow[f]=10
        user1[f] = 20
        user2[f] = 0.5*(f-30)
        user3[f] = 0.5*(f-30)
    elif f<100:
        env_flow[f]=10
        user1[f] = 20
        user2[f] = 30+f-90
        user3[f] = 30
    else:
        env_flow[f]=10+f-100
        user1[f] = 20
        user2[f] = 40
        user3[f] = 30
```

```python
import matplotlib.pyplot as plt

plt.plot(gage_flow,env_flow, label = "Env flow")
plt.plot(gage_flow,user1, label = "User 1")
plt.plot(gage_flow,user2, label = "User 2")
plt.plot(gage_flow,user3, label = "User 3")
plt.xlabel("Flow at the gage site")
plt.ylabel("Operating policy for each user")
plt.legend()
```

```python
def operating_policy(f):
    return env_flow[f], user1[f], user2[f], user3[f]

print(operating_policy(120))
```

```python
#Lecture 06
#Simulate for each possible flow
#Initialize variables
env_flow = np.zeros(130)
user1 = np.zeros(130)
user2 = np.zeros(130)
user3 = np.zeros(130)

f01 = np.zeros(130)
f12 = np.zeros(130)
f23 = np.zeros(130)

s1 = np.zeros(130)
s2 = np.zeros(130)
s3= np.zeros(130)


for f in range(130):
    if f <10:
        f01[f] = f
        s1[f] = 0
        f12[f] = f - s1[f]*(1-0.6)
        s2[f] = 0
        f23[f] = f12[f] - s2[f]*(1-0.6)
        s3[f] = 0
        env_flow[f] = f23[f] - s3[f]*(1-0.2)
        
    elif f<30:
        f01[f] = f
        s1[f] = 0
        f12[f] = f - s1[f]*(1-0.6)
        s2[f] = 0
        f23[f] = f12[f] - s2[f]*(1-0.6)
        s3[f] = f - 10
        env_flow[f] = f23[f] - s3[f]*(1-0.2)
        
        

    elif f<90:
        f01[f] = f
        s1[f] = 0.5*(f-30)
        f12[f] = f - s1[f]*(1-0.6)
        s2[f] = 0.5*(f-30)
        f23[f] = f12[f] - s2[f]*(1-0.6)
        s3[f] = 20
        env_flow[f] = f23[f] - s3[f]*(1-0.2)
        
        
    elif f<100:
        
        f01[f] = f
        s1[f] = 30
        f12[f] = f - s1[f]*(1-0.6)
        s2[f] = 30+f-90
        f23[f] = f12[f] - s2[f]*(1-0.6)
        s3[f] = 20
        env_flow[f] = f23[f] - s3[f]*(1-0.2)
        
        
    else:
        
        f01[f] = f
        s1[f] = 30
        f12[f] = f - s1[f]*(1-0.6)
        s2[f] = 40
        f23[f] = f12[f] - s2[f]*(1-0.6)
        s3[f] = 20
        env_flow[f] = f23[f] - s3[f]*(1-0.2)
        
import matplotlib.pyplot as plt

plt.plot(gage_flow,env_flow, label = "Env flow")
plt.plot(gage_flow,s1, label = "User 1")
plt.plot(gage_flow,s2, label = "User 2")
plt.plot(gage_flow,s3, label = "User 3")
plt.xlabel("Flow at the gage site")
plt.ylabel("Operating policy for each user")
plt.legend()
```
