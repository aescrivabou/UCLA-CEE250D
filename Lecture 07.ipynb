## Exercise 4.9 in Loucks and van Beek (2017) - Lagrange multiplier

```python
import cvxpy as cvx


# Variables
x1 = cvx.Variable(name="x1")
x2 = cvx.Variable(name="x2")
x3 = cvx.Variable(name="x3")

# Objective function
obj = cvx.Maximize(12*x1-x1**2 + 8*x2 - x2**2 + 18*x3 - 3*x3**2)

#Constraint
constraints = [x1 + x2 + x3 == 10.1]

#Solving the problem
prob = cvx.Problem(obj, constraints)
prob.solve()

#Printing the results
print('Objective = %f' % obj.value)
print('x1 = %f' % x1.value)
print('x2 = %f' % x2.value)
print('x3 = %f' % x3.value)

for c in constraints:
  print('Dual (%s) = %f' % (c, c.dual_value))


```

## Complex functions to show local vs global minimum/maximum

```python
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# fig = plt.figure()
# ax = plt.axes(projection='3d')

def f(x, y):
    return x*y*np.exp(-x**2 - y**2)

x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
ax.set_title('surface')
```
